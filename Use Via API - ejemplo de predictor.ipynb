# Prueba 0 - Usando parametros de default con hugging face

```python

from gradio_client import Client

client = Client("mlgarcia/edvai")
result = client.predict(
		param_0="Casa",
		param_1="Capital Federal",
		param_2="Palermo",
		param_3=2,
		param_4=1,
		param_5=1,
		param_6=60,
		param_7=50,
		api_name="/predict_price"
)
print(result)
```
